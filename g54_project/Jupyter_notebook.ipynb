```python
import pandas as pd
import sqlite3

df = pd.read_csv("data/G54_data.csv")

df = df.drop(columns=["institution_id", "institution_name", "institution_created_date", "conference_id", "conference_date", "conference_comments", "registration_id",
"registration_fee", "attendance_status", "conference_title"])

df = df.drop_duplicates(subset=["researcher_id"], keep="first")

conn = sqlite3.connect("data/Bases_dados.db")

df.to_sql("Researcher", conn, if_exists = "replace", index = False)

conn.close()
print("Sucesso! Agora podes abrir o ficheiro .db no DB Browser.")
```
